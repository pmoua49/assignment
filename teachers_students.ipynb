#### Format txt files into Dataframes 

```python
df_reference = pd.read_csv('IndicatorReference.txt', sep='|')
df_reference
```

```python
df_results = pd.read_csv('Results.txt', sep='|')
df_results
```

```python
df_results["SchoolCode"].value_counts()
```

```python
df_school_ref = pd.read_csv('SchoolReference.txt', sep="|")
df_school_ref
```

```python
df_results.info()
```

```python
df_school_ref.info()
```

```python
results1 = df_results.astype({"academicyear":'object'})
school_ref1 = df_school_ref.astype({"academicyear":'object'})
```

```python
results1.info()
```

```python
#concat dfs with keys 
joined_df = pd.merge(results1, school_ref1, how='left', on=["SchoolCode", "academicyear"])
joined_df
```

```python
joined_df.isnull().sum()
#will use variables that have 0-1 nulls 
```

```python
new_df = joined_df[['indicator', 'academicyear', 'District_x', 'School_x', 'subgroup', 'Denominator', 'PercentageStudents', 'RoadMapRegionDenominator', 'RoadMapRegionPercentage']]
```

```python
new_df.dropna(inplace=True)
```

```python
new_df.info()
```
